```python
import random


def get_movie():
    movies = ['Scarface', 
              'Eyes Wide Shut']
    return random.choice(movies).upper()

class BasicHints:
    
     def __init__(self, country, genre, director, starring):
            self.country = country
            self.genre = genre
            self.director = director
            self.starring = starring

scarface = BasicHints('United States', 'Crime', 'Brian de Palma', 'Al Pacino')
eyeswideshut = BasicHints('United States', 'Mystery', 'Stanley Kubrick', 'Tom Cruise')

def additional_hint(question):
    ynanswer = input(question).upper()
    if ynanswer =='Y' or ynanswer =='N':
        return ynanswer
    else:
        print('Invalid input, please try again.')
        return additional_hint(question)
    
scarface_additional_hints = ['A Cuban immigrant takes over a drug cartel.', 
                             'The movie portrays Miami in the 80\'s.',
                             'The movie was a major inspiration for the 2002 video game Grand Theft Auto: Vice City.',
                             'Tony Montana\'s most remarkable line: "Say hello to my little friend".']
eyeswideshut_additional_hints = ['Director Stanley Kubrick died six days after showing the final movie cut',
                                'Tom Cruise and Nicole Kidman played the cuple Dr.\"Bill\" Harford and Alice Harford while married in real life.',
                                'The film\'s production, at 400 days, holds the Guinness World Record for the longest continuous film shoot',
                                'The film was described as an erotic thriller']

def check(movie, guesses, guess):
    score = 100
    status = ''
    matches = 0
    for letter in movie:
        if letter in guesses:
            status += letter
        else:
            status += '_'
            
        if letter == guess:
            matches += 1
            
    if matches > 1:
        print('Yes! The movie contains',matches,'"' + guess + '"' + 's')
        answer = additional_hint("Would you like a new hint? \n(Y/N)")
        if answer == 'Y':
            print('a')
        elif answer == 'N':
            pass
            
    elif matches == 1:
        print('Yes! The movie contains',matches, '"' + guess + '"')
        answer = additional_hint("Would you like a new hint? \n(Y/N)")
        if answer == 'Y':
            print('a')
        elif answer == 'N':
            pass
  
    else:
        print('Sorry. The movie does not contain the letter "' + guess + '"')
        answer = additional_hint("Would you like a new hint? \n(Y/N)")
        if answer == 'Y':
            print('a')
        elif answer == 'N':
            pass

    return status

def main():
    movie = get_movie()
    guesses = []
    guessed = False
    print('The movie contains', len(movie), 'letters.')
    if movie == 'SCARFACE':
        print('Your starting hints are:','\n\nCountry:', scarface.country,'\nGenre:',scarface.genre,'\nDirector:',scarface.director,'\nStarring:',scarface.starring)
    elif movie == 'EYES WIDE SHUT':
        print('Your starting hints are:','\n\nCountry:', eyeswideshut.country,'\nGenre:',eyeswideshut.genre,'\nDirector:',eyeswideshut.director,'\nStarring:',eyeswideshut.starring)
   
    while not guessed:
        text = 'Please enter one letter or a {}-letter movie. '.format(len(movie))
        guess = input(text)
        guess = guess.upper()
        
        if guess in guesses:
            print('You already guessed "' + guess + '"')
            answer = additional_hint("Would you like a new hint? \n(Y/N)")
            if answer == 'Y':
                 print('a')
            elif answer == 'N':
                pass
        
        elif len(guess) == len(movie):
            guesses.append(guess)
            if guess == movie:
                guessed = True
            else:
                print('Sorry, this is not the movie!')
                answer = additional_hint("Would you like a new hint? \n(Y/N)")
                if answer == 'Y':
                     print('a')
                elif answer == 'N':
                    pass
                
        elif len(guess) == 1:
            guesses.append(guess)
            result = check(movie,guesses,guess)
            if result == movie:
                guessed = True
            else:
                print(result)   
                
        else:
            print('Invalid entry.')
            
    print('Yes, the movie is',movie,'! You got in', len(guesses),'tries.')
    
main()
```
