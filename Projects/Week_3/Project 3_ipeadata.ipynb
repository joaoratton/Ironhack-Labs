```python
df = pd.read_html(html)[2]
df
```

```python
df1 = df[793:]
data = df1.rename({0:'year', 1:'ex.rate'}, axis=1)
data['ex.rate'] = data['ex.rate'].astype(np.int64)
data['ex.rate'] = data['ex.rate']/100
```

```python
engine = db.create_engine('postgresql+psycopg2://postgres:180592@localhost/project_week3')
conn = engine.connect()
```

```python
data.to_sql('exchange_rate', con=conn, if_exists='replace', index=False)
```
